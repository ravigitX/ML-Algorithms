```python
# load the iris dataset
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report


# Load and store the feature matrix (X) and response vector (y)
iris = load_iris()
X = iris.data
y = iris.target

#print(iris)

print("\nFeature Matrix :\n",X[:10,:])
print("\nTarget Vector : \n",y[:10,])

# splitting X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# training the model on training set
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# making predictions on the testing set
y_pred = gnb.predict(X_test)

for i in range(len(X_test)):
        print("\n",X_test[i],"-->", "(Predicted Values)",y_pred[i], "(Actual values)",y_test[i])

# comparing actual response values (y_test) with predicted response values (y_pred)
#print("\n Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
def cal_accuracy(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    print("\n\nConfusion Matrix: \n", cm)
    print("\nAccuracy: ", accuracy_score(y_test, y_pred) * 100)
    print("\nReport: \n", classification_report(y_test, y_pred))
    # Plotting the confusion matrix heatmap
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix Heatmap')
    plt.show()

cal_accuracy(y_test, y_pred)
```
